```python
import numpy as np
```

```python
f = open("./input", "r")
lines = f.read().strip().splitlines()
```

```python
grid_s = np.array([[x for x in l] for l in lines], dtype=str)
grid = np.zeros(grid_s.shape, dtype=int)
grid[np.where(grid_s == "^")] = 1
```

```python
beams = np.zeros(grid.shape[1],dtype=int)
beams[np.where(grid_s[0] == "S")] = 1
```

```python
part1 = 0
curr_beam = beams.copy()
for row in grid:
    new_beam = curr_beam - row
    new_beam[np.where(new_beam < 0)] = 0
    diff_beam = curr_beam - new_beam
    for i in np.where(diff_beam == 1)[0]:
        part1 += 1
        new_beam[[i-1, i+1]] = 1
    curr_beam = new_beam
part1
```

```python
from collections import defaultdict

curr_beam = beams.copy()
mem_map = defaultdict(int)
mem_map[np.where(beams == 1)[0][0]] += 1

for row in grid:
    new_beam = curr_beam - row
    new_beam[np.where(new_beam < 0)] = 0
    diff_beam = curr_beam - new_beam
    for i in np.where(diff_beam == 1)[0]:
        ways = mem_map[i]
        mem_map[i - 1] += ways
        mem_map[i + 1] += ways
        mem_map[i] = 0
        new_beam[[i-1, i+1]] = 1
    curr_beam = new_beam

sum(mem_map.values())
```
