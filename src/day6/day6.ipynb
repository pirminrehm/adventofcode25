```python
import numpy as np
from operator import add, mul
from functools import reduce
```

```python
f = open("./input", "r")
lines = f.read().splitlines()
```

```python
nrs = np.array([list(map(int,l.split())) for l in lines[:-1]])
args = lines[-1].split()
nrs_t = nrs.T
args_f = [mul if a == "*" else add for a in args]
```

```python
part1 = 0
for i in range(len(nrs_t)):
    part1 += reduce(args_f[i], nrs_t[i])
part1
```

```python
nrs_cs = np.array([[x for x in l] for l in lines[:-1]], dtype=str).T
nrs_cr = [reduce(lambda x1, x2: x1+ x2 ,n).strip() for n in nrs_cs]
nrs_c = [[]]
for n in nrs_cr:
    if n == "":
        nrs_c.append([])
        continue
    nrs_c[-1].append(int(n))
```

```python
part2 = 0
for i in range(len(nrs_c)):
    part2 += reduce(args_f[i], nrs_c[i])
part2
```
