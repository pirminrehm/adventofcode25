```python
import numpy as np
import re
```

```python
f = open("./input", "r")
lines = f.read().strip().splitlines()
```

```python
turns = []
for l in lines:
    nr = int(l[1:])
    if l[0] == "L":
        nr *= -1
    turns.append(nr)
```

```python
curr = 50
part1 = 0

for turn in turns:
    curr = (curr + turn) % 100
    if curr == 0:
        part1 += 1

part1
```

```python
curr = 50
part2 = 0

for turn in turns:
    turnovers = (curr + turn) // 100
    nxt = (curr + turn) % 100
    if curr == 0 and turnovers < 0:  # do not double count if going left when already on 0
        turnovers += 1
    abs_turnovers = abs(turnovers)
    if nxt == 0 and turn < 0:  # count 0 if going back from pos int to 0
        abs_turnovers += 1
    part2 += abs_turnovers
    curr = nxt

part2
```
