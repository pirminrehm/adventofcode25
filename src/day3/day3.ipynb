```python
import numpy as np
from itertools import pairwise
```

```python
f = open("./input", "r")
lines = f.read().strip().splitlines()
```

```python
banks = [[int(x) for x in l] for l in lines]
```

```python
part1 = 0
for bank in banks:
   m1 = max(bank)
   m1i = bank.index(m1)
   rest = bank[m1i+1:]
   if len(rest) == 0:
      m1 = max(bank[:m1i])
      m1i = bank.index(m1)
   m2 = max(bank[m1i+1:])
   part1 += int(str(m1) + str(m2))

part1
```

```python
def find_max_in_range(lst, start, end):
    l = lst[start:end]
    m = max(l)
    i = l.index(m)
    return m, i
```

```python
part2 = 0
for bank in banks:
    lower = 0
    nr = ""
    for upper in range(11,-1,-1):
        if upper == 0:
            upper = -len(bank)
        m, i = find_max_in_range(bank, lower, -upper)
        lower = lower + i + 1
        nr += str(m)
    part2 += int(nr)
part2
```
