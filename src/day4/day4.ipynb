```python
import numpy as np
from itertools import pairwise
```

```python
f = open("./input", "r")
lines = f.read().strip().splitlines()
```

```python
grid = np.array([[x for x in l] for l in lines], dtype=str)
```

```python
from itertools import permutations

adjacents = list(permutations((0, 1, -1), 2))
adjacents.extend([(1,1),(-1,-1)])
adjacents = sorted(adjacents)
adjacents
```

```python
def t_add(t1, t2):
    return (t1[0] + t2[0], t1[1] + t2[1])
```

```python
part1 = 0
rows_len = len(grid)
cols_len = len(grid[0])
for ri in range(rows_len):
    row = grid[ri]
    for ci in range(cols_len):
        point = row[ci]
        if point != "@":
            continue
        paper_count = 0
        for neighbour in [t_add(x, (ri, ci)) for x in adjacents]:
            if neighbour[0] < 0 or neighbour[0] >= rows_len or neighbour[1] < 0 or neighbour[1] >= cols_len:
                continue
            if grid[neighbour] == "@":
                paper_count += 1
        if paper_count < 4:
            part1 += 1
part1

```

```python
g2 = grid.copy()
part2 = 0
rows_len = len(g2)
cols_len = len(g2[0])
papers_found = 1

while papers_found != 0:
    papers_found = 0
    for ri in range(rows_len):
        row = g2[ri]
        for ci in range(cols_len):
            point = row[ci]
            if point != "@":
                continue
            paper_count = 0
            for neighbour in [t_add(x, (ri, ci)) for x in adjacents]:
                if neighbour[0] < 0 or neighbour[0] >= rows_len or neighbour[1] < 0 or neighbour[1] >= cols_len:
                    continue
                if g2[neighbour] == "@":
                    paper_count += 1
            if paper_count < 4:
                papers_found += 1
                g2[(ri, ci)] = "x"
    part2 += papers_found
part2
```
