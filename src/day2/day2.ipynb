```python
import numpy as np
import re
```

```python
f = open("./input", "r")
lines = f.read().strip().splitlines()
```

```python
ranges = [list(map(int,l.split("-"))) for l in lines[0].split(",")]
```

```python
step = "119119"
a  = step[:len(step)//2]
b = step[len(step)//2:]
a == b
```

```python
part1 = 0
for r in ranges:
    for i in range(r[0], r[1]+1, 1):
        step = str(i)
        if len(step)%2: # odd length
            continue
        a  = step[:len(step)//2]
        b = step[len(step)//2:]
        if a == b:
            part1 += i
part1
```

```python
part2 = 0
for r in ranges:
    for i in range(r[0], r[1]+1, 1):
        if re.fullmatch(r'(\d+)\1+', str(i)):
            part2 += i
part2
```
