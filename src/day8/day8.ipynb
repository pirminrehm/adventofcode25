```python
import numpy as np
import math
```

```python
f = open("./input", "r")
lines = f.read().strip().splitlines()
```

```python
boxes = sorted([list(map(int, x.split(","))) for x in lines])
```

```python
distances = []

for i in range(len(boxes)):
    for n in range(i+1, len(boxes),1):
        dist = math.dist(boxes[i], boxes[n])
        distances.append((dist, (boxes[i], boxes[n])))

distances = sorted(distances)
```

```python
next_circuit_id = 0
circuit_to_box = {}

for i in range(1000):
    d1, d2 = distances[i][1]
    for cid, c in circuit_to_box.items():
        d1_in = d1 in c
        d2_in = d2 in c

        if not d1_in and not d2_in:
            continue
        if d1_in and d2_in:
            break

        if d2_in:
            d1, d2 = d2, d1 # now d1 is always in

        # check if already in circuit
        for cid_s, c_s in circuit_to_box.items():
            if cid == cid_s:
                continue
            if d2 in c_s:
                c.extend(c_s)
                circuit_to_box[cid_s] = []
                break
        else:
            c.append(d2)
        break
    else:
        circuit_to_box[next_circuit_id] = [d1, d2]
        next_circuit_id += 1
```

```python
c_length = sorted([len(x) for x in circuit_to_box.values()],reverse=True)

part1 = c_length[0] * c_length[1] * c_length[2]
part1
```

```python
next_circuit_id = 0
circuit_to_box = {}

i = 0
last_junction = None
while i >= 0:
    d1, d2 = distances[i][1]
    for cid, c in circuit_to_box.items():
        d1_in = d1 in c
        d2_in = d2 in c

        if not d1_in and not d2_in:
            continue
        if d1_in and d2_in:
            break

        if d2_in:
            d1, d2 = d2, d1 # now d1 is always in

        # check if already in circuit
        for cid_s, c_s in circuit_to_box.items():
            if cid == cid_s:
                continue
            if d2 in c_s:
                c.extend(c_s)
                circuit_to_box[cid_s] = []
                break
        else:
            c.append(d2)

        if len(c) == len(boxes):
            last_junction = [d1, d2]
            i = -10
        break
    else:
        circuit_to_box[next_circuit_id] = [d1, d2]
        next_circuit_id += 1
    
    i += 1
```
