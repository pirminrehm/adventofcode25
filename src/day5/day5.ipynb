```python
import numpy as np
```

```python
f = open("./input", "r")
fi = f.read().strip()
```

```python
_ranges, _ids = tuple([x.split("\n") for x in fi.split("\n\n")])
ranges = [list(map(int,x.split("-"))) for x in _ranges]
ranges = sorted(ranges)
ids = list(map(int, _ids))
```

```python
part1 = 0
for nr in ids:
    for rc in ranges:
        if nr < rc[0]:
            break
        if nr > rc[1]:
            continue
        part1 += 1
        break
part1
```

```python
final_r = [ranges[0]]
curr_i = 0
for i in range(1, len(ranges), 1):
    rc = final_r[curr_i]
    rn = ranges[i]

    if rc[1] >= rn[0]:
        uppper = rn[1]
        if rc[1] > uppper:
            uppper = rc[1]
        final_r[curr_i] = [rc[0], uppper]
        continue

    curr_i += 1
    final_r.append(rn)

part2 = 0
for r in final_r:
    part2 += r[1] - r[0] + 1
part2
```
